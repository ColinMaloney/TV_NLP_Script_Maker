```python
randm_corpus = []
for s in range(1,4):
    for ep in range(1,12):
        try:
            url = f'https://www.springfieldspringfield.co.uk/view_episode_scripts.php?tv-show=rick-and-morty-2013&episode=s0{s}e0{ep}'
            html = requests.get(url, headers=headers)
            soup = BeautifulSoup(html.text, 'html.parser')
            script = soup.find_all(class_=re.compile("scrolling"))
            randm_corpus.append(script)
            time.sleep(5)
        except: 
            print(f'No ep:{ep}')
```
