```python
Sherlock = f'Sherlock_s4_ep2'
```

```python
PDF_file = f"Sherlock/{Sherlock}.pdf"
```

```python
pages = convert_from_path(PDF_file, 68) 
```

```python
image_counter = 1
for page in pages: 
    filename = "page_"+str(image_counter)+".jpg"
    page.save(filename, 'JPEG') 
    image_counter = image_counter + 1
```

```python
filelimit = image_counter-1
outfile = f"{Sherlock}.txt"
f = open(outfile, "a") 
for i in range(1, filelimit + 1): 
    filename = "page_"+str(i)+".jpg"
    text = str(((pytesseract.image_to_string(Image.open(filename))))) 
    text = text.replace('-\n', '')     
    f.write(text) 
f.close() 
```

```python
with open(f'{Sherlock}.txt')as f:
    for line in f:
        print(line)
```
